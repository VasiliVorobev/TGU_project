```python

# 6. Словари
# Задача: "Кулинарная книга"


pancakes1 = ('мука,яйца,молоко,сахар,масло')
salad_caesar1 = ('салат,курица,сухарики,сыр,соус')
chokolate_cookiees1 = ('мука,шоколад,масло,сахар,яйца')

cookbook = {"Блины":30,"Салат Цезарь":20,"Шоколадное печенье":45}


# решение
print("=== ВСЕ РЕЦЕПТЫ ===")
print()
for index,x in enumerate(cookbook):
    print(f"{index+1}. {x}")
print()
print("=== ДЕТАЛИ РЕЦЕПТОВ ===")
print()
for x,y in   cookbook.items():
    print(f"{x}:")
    if x == 'Блины':
        print(f"    Ингредиенты: {pancakes1}")
    elif x == 'Салат Цезарь':
        print(f"    Ингредиенты: {salad_caesar1}")
    elif x == 'Шоколадное печенье':
        print(f"    Ингредиенты: {chokolate_cookiees1}")
        print(f"    Время: {y} минут")
print()
for x,y in   cookbook.items():
    if y == min(cookbook.values()):
        print (f"Самый быстрый: {x} ({y} минут)")
    elif y == max(cookbook.values()):
        print (f"Самый долгий: {x} ({y} минут)")

```

```python
# 9.
# Задача: "Преобразование операций с текстом"

print("УМНЫЙ КАЛЬКУЛЯТОР ОПЕРАЦИЙ С ТЕКСТОМ")
print()
print("Доступные команды:")
print("- uppercase: преобразовать в верхний регистр")
print("- lowercase: преобразовать в нижний регистр")
print("- length: посчитать длину текста")
print("- words: посчитать количество")
print()
command = input("Введите команду (uppercase, lowercase, length, words, reverse): ")
text = input("Введите текст: ")

if command == 'uppercase':
    result = text.upper()
    print()
    print(f"Результат:{result}")
elif command == 'lowercase':
    result = text.lower()
    print()
    print(f"Результат:{result}")
elif command == 'length':
    length = len(text)
    print()
    print(f"Длина текста: {length} символов")
elif command == 'words':
    words_count = len(text.split())
    print()
    print(f"Количество слов: {words_count}")
elif command == 'reverse':
    reversed_text = text[::-1]
    print()
    print(f"Перевернутый текст:{reversed_text}")
else:
    print()
    print("Неизвестная команда.")
```
